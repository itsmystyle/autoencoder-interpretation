```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
def plot_learning(path):
    with open(path, 'r') as fin:
        js = json.loads(fin.read())

    tl = [l['loss'] for l in js['train']]
    vl = [l['loss'] for l in js['valid']]
    
    tr = [l['Accuracy'] for l in js['train']]
    vr = [l['Accuracy'] for l in js['valid']]
    
    fig = plt.figure(figsize = (10,8))
    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(222)
    
    fig.suptitle('Learning Curve', fontsize=16)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    ax1.set_xlabel('epochs')
    ax1.set_ylabel('loss')
    ax1.plot(tl, label='train')
    ax1.plot(vl, label='valid')
    ax1.legend()
    
    ax2.set_xlabel('epochs')
    ax2.set_ylabel('accuracy')
    ax2.plot(tr, label='train')
    ax2.plot(vr, label='valid')
    ax2.legend()
```

```python
target = 'models/hn_32/log.json'
```

```python
t_1 = plot_learning(target)
```
