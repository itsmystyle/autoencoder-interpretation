```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
def plot_learning(path):
    with open(path, 'r') as fin:
        js = json.loads(fin.read())

    tl = [l['loss'] for l in js['train']]
    vl = [l['loss'] for l in js['valid']]
    tsl = [l['loss'] for l in js['test']]
    
    tr = [l['Accuracy'] for l in js['train']]
    vr = [l['Accuracy'] for l in js['valid']]
    tsr = [l['Accuracy'] for l in js['test']]
    
    tf = [l['F1_micro'] for l in js['train']]
    vf = [l['F1_micro'] for l in js['valid']]
    tsf = [l['F1_micro'] for l in js['test']]
    
    tp = [l['F1_macro'] for l in js['train']]
    vp = [l['F1_macro'] for l in js['valid']]
    tsp = [l['F1_macro'] for l in js['test']]
    
    ta = [l['F1_weighted'] for l in js['train']]
    va = [l['F1_weighted'] for l in js['valid']]
    tsa = [l['F1_weighted'] for l in js['test']]
    
    plt.title('Loss')
    plt.plot(tl, label='train')
    plt.plot(vl, label='valid')
    plt.plot(tsl, label='test')
    plt.legend()
    plt.show()
    
    fig = plt.figure(figsize = (12,8))
    ax1 = fig.add_subplot(221)
    ax2 = fig.add_subplot(222)
    ax3 = fig.add_subplot(223)
    ax4 = fig.add_subplot(224)
    
    ax1.set_title('Accuracy')
    ax1.plot(tr, label='train')
    ax1.plot(vr, label='valid')
    ax1.plot(tsr, label='test')
    ax1.legend()
              
    ax2.set_title('F1 Macro')
    ax2.plot(tp, label='train')
    ax2.plot(vp, label='valid')
    ax2.plot(tsp, label='test')
    ax2.legend()
    
    ax3.set_title('F1 Weighted')
    ax3.plot(ta, label='train')
    ax3.plot(va, label='valid')
    ax3.plot(tsa, label='test')
    ax3.legend()
    
    ax4.set_title('F1 Micro')
    ax4.plot(tf, label='train')
    ax4.plot(vf, label='valid')
    ax4.plot(tsf, label='test')
    ax4.legend()
```

```python
target = 'models/hnn_v5/log.json'
```

```python
t_1 = plot_learning(target)
```
